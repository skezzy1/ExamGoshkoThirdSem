```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('tips.csv', header=None, names=['total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size', 'tip_percentage'])

df['tip'] = pd.to_numeric(df['tip'], errors='coerce')
```

```python
df = df[df['smoker'].isin(['Yes', 'No'])]
df['smoker'] = df['smoker'].replace({'Yes': 'Так', 'No': 'Ні'})

average_tips = df.groupby('smoker')['tip'].mean()
print(average_tips)

average_tips.plot(kind='bar')

plt.title('Хто дає більші чайові: курці чи ні')
plt.xlabel('Чи курець')
plt.ylabel('Середнє арифметичне')

plt.show()
```

```python
df['day'] = df['day'].replace({'Fri': "П'ятниця", 'Sat': 'Субота', 'Sun': 'Неділя', 'Thur': 'Четвер'})
average_tips = df.groupby('day')['tip'].mean()
print(average_tips)

average_tips.plot(kind='bar')

plt.title('Чайові по днях тижня')
plt.xlabel('День тижня')
plt.ylabel('Середнє арифметичне')

plt.show()
```

```python
df['time'] = df['time'].replace({'Dinner': 'Вечеря', 'Lunch': 'Обід'})
average_tips = df.groupby('time')['tip'].mean()
print(average_tips)

average_tips.plot(kind='bar')

plt.title('Чайові в залежності від часу')
plt.xlabel('Година дня')
plt.ylabel('Середнє арифметичне')

plt.show()
```

```python
df['total_bill'] = pd.to_numeric(df['total_bill'], errors='coerce')

plt.hist(df['total_bill'], bins=8, range=[0, 100], edgecolor='black')

plt.title('Гістограма рахунків')
plt.xlabel('Розподіл рахунків')
plt.ylabel('Частота появи')

plt.show()
```

```python
plt.scatter(df['total_bill'], df['tip'])

plt.title('Скаттер рахунок-чайові')
plt.xlabel('Рахунок')
plt.ylabel('Чайові')

plt.show()
```

```python
df['sex'] = df['sex'].replace({'Female': 'Жінка', 'Male': 'Чоловік'})
average_tips = df.groupby('sex')['tip'].mean()
print(average_tips)

average_tips.plot(kind='bar')

plt.title('Залежність чайових від статі клієнта')
plt.xlabel('Стать')
plt.ylabel('Середнє арифметичне')

plt.show()
```
