```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class TitanicData:
    def __init__(self, excel_file):
        self.data = pd.read_excel(excel_file)

    def clean_data(self):
        self.data = self.data[pd.to_numeric(self.data['Age'], errors='coerce').notna()]

    def plot_survived_age_distribution(self, gender):
        survived_data = self.data[(self.data['Survived'] == 1) & (self.data['Sex'] == gender)]

        plt.hist(survived_data['Age'], bins=20, color='pink' if gender == 'female' else 'blue',
                 edgecolor='black', alpha=0.7, label=f'Survived {gender}s')
        plt.title(f'Age distribution of survived {gender}s')
        plt.xlabel('Age')
        plt.ylabel('Count')
        plt.legend()
        plt.show()
    def plot_pclass_pie_chart(self):
        pclass_counts = self.data['Pclass'].value_counts()
        
        plt.pie(pclass_counts, labels=pclass_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'lightcoral'])
        plt.title('Passenger class distribution')
        plt.show()
```

```python
analyzer = TitanicData('stinky_titanic_data.xlsx')
analyzer.clean_data()
```

```python
analyzer.plot_survived_age_distribution(gender='female')
```

```python
analyzer.plot_survived_age_distribution(gender='male')
```

```python
analyzer.plot_pclass_pie_chart()
```
